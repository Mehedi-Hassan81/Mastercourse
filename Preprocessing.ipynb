```python
df1 = pd.read_csv("renewable_twh.csv")
df2 = pd.read_csv("renewable_percent.csv")
```

```python
df1 = df1.drop(columns="Year")
df2 = df2.drop(columns="Year")
```

```python
print(df1)
print(df2)
```

```python
twh_dict = {
    "Renew": "Renew_twh",
    "Hydro": "Hydro_twh", 
    "Wind": "Wind_twh",
    "Solar": "Solar_twh",
    "Bio": "Bio_twh",
    "Geo": "Geo_twh",
}
```

```python
percent_dict = {
    "Renew": "Renew_percent",
    "Hydro": "Hydro_percent", 
    "Wind": "Wind_percent",
    "Solar": "Solar_percent",
    "Bio": "Bio_percent",
    "Geo": "Geo_percent",
}
```

```python
df1
```

```python
df1 = df1.rename(columns=twh_dict)
```

```python
df1
```

```python
df2 = df2.rename(columns=percent_dict)
```

```python
df2
```

```python
df1 = df1[df1["Location"] != "World"]
df1 = df1.reset_index(drop=True)
```

```python
df1.head()
```

```python
df1.describe
```

```python
df2 = df2[df2["Location"] != "World"]
df2 = df2.reset_index(drop=True)
```

```python
df2.head()
```

```python
df2.describe
```

```python
same_values = set(df1["Location"].unique()) == set(df2["Location"].unique())
print(same_values)
```

```python
final_df = pd.merge(df1, df2, on="Location", how="inner")
```

```python
final_df.describe
```

```python
final_df.head()
```

```python
final_df.to_csv("Renewable_Energy_Production.csv", index=False)
```
